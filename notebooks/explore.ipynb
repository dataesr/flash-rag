```python
import httpx


def fetch_data(url: str):
    with httpx.Client() as client:
        try:
            response = client.get(url)
            response.raise_for_status()
            return response.json()
        except Exception as error:
            print(f"Error while fetching {url=}: {error}")


def get_data():
    base_url = "https://zenodo.org/api/records?communities=ssm-esr&size=25&page=1"
    data = fetch_data(base_url)
    return data
```

```python
import os
from mistralai.client import Mistral
from dotenv import load_dotenv

load_dotenv()

client = Mistral(api_key=os.getenv("MISTRAL_API_KEY"))
```

```python
import base64


def encode_pdf(pdf_path):
    try:
        with open(pdf_path, "rb") as pdf_file:
            return base64.b64encode(pdf_file.read()).decode("utf-8")
    except Exception as error:
        print(f"Error: {error}")
        return None
```

```python
import json

pdf_path = "./data/pdf/1003.pdf"
base64_pdf = encode_pdf(pdf_path)

response = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "document_url",
        "document_url": f"data:application/pdf;base64,{base64_pdf}",
        "document_name": "1003.pdf",
    },
    extract_footer=True,
    extract_header=True,
)
data = json.loads(response.model_dump_json())
```

```python
import re
from dataclasses import dataclass, field


def table_to_dict(raw: str):
    lines = [line.strip() for line in raw.strip().splitlines()]
    # Remove separator line
    lines = [line for line in lines if not re.match(r"^\|[-:\s|]+\|$", line)]

    rows = []
    for line in lines:
        cells = [cell.strip() for cell in line.strip("|").split("|")]
        rows.append(cells)

    if not rows:
        return None

    headers = rows[0]
    data = rows[1:]

    return {"headers": headers, "data": data}


@dataclass
class Section:
    level: int  # 1=title, 2=h2, 3=h3
    title: str
    paragraphs: list[str] = field(default_factory=list)
    tables: list[str] = field(default_factory=list)  # raw markdown table strings


def parse_markdown(md: str) -> list[Section]:
    sections: list[Section] = []
    current: Section | None = None
    buffer: list[str] = []

    def flush_buffer():
        if current is None or not buffer:
            return
        block = "\n".join(buffer).strip()
        if not block:
            return

        # Split block into chunks separated by blank lines
        for chunk in re.split(r"\n{2,}", block):
            chunk = chunk.strip()
            if not chunk:
                continue

            lines = chunk.splitlines()
            # A table has at least one | and a separator line with ---
            is_table = any("|" in line for line in lines) and any(re.match(r"^\s*\|[\s\-|:]+\|\s*$", line) for line in lines)

            if is_table:
                current.tables.append(table_to_dict(chunk))
            else:
                if chunk:
                    current.paragraphs.append(chunk)
        buffer.clear()

    for line in md.splitlines():
        heading = re.match(r"^(#{1,3})\s+(.*)", line)
        if heading:
            flush_buffer()
            level = len(heading.group(1))
            title = heading.group(2).strip()
            current = Section(level=level, title=title)
            sections.append(current)
        else:
            if current is not None:
                buffer.append(line)

    flush_buffer()
    return sections
```
