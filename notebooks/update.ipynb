```python
import os
import httpx
from dotenv import load_dotenv

load_dotenv()
```

```python
headers = {"Authorization": os.getenv("FLASH_RAG_API_KEY", "")}
url = "https://flash-rag.staging.dataesr.ovh/update"
```

```python
# update flash rag documents
payload = {
    "load_skip_fetch": True,
    "load_skip_download": False,
    "load_force_download": False,
    "extract_force": False,
    "parse_force": False,
    "populate_override": True,
    "populate_reset": False,
}
response = httpx.post(url, headers=headers, json=payload)
response
```
